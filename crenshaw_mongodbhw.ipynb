```python
#$AND/$ALL Operation
```

```python
#1 Write a MongoDB query to display any 5 movies with both the genres: “Adventure” and “Movie” in collection movies (use $all)
for doc in db.movies.find({"genres":{"$all":["Adventure","Movie"]}}).limit(5):
    pprint(doc)
#there are no movies that have the genre "Movie" listed. There are many movies, however, with the "Adventure" genre tag. 
#this was confirmed by searching directly in the MongoDBCompass platform
```

```python
#2 Write a MongoDB query to display any 5 movies with both the condition: genres “Adventure” and cast “Tom Hanks”.
for doc in db.movies.find({"$and":[{"genres":{"$all":["Adventure"]}}, {"cast":{"$all":["Tom Hanks"]}}]}).limit(5):
    pprint(doc)
```

```python
#Aggregation
```

```python
#1 Write a MongoDB query to display average number of awards won by a movie (use aggregate function with $avg operator).
#this code displays the average number of awards won by movies across the collection. 
#finding the average of one movie will simply result in the number of awards won (no true "average" is completed)
for doc in db.movies.aggregate([
    {"$group": {"_id": "", "average_award_wins": { "$avg": "$awards.wins" }}}, 
      { "$project" : { "_id" : 0,}}]):
    pprint(doc)
```

```python
#2 Write a MongoDB query to display most awards won by a movie (use aggregate function with $max operator).
#12 Years a Slave won the most awards. 
for doc in db.movies.aggregate( [{ "$group": {"_id": "$title", "maxnumofawards": { "$max":"$awards.wins" } } }, 
                                 { "$sort" : { "maxnumofawards" : -1 } },
                                {"$limit":1}] ):
    pprint(doc)
```
